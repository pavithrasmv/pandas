```python
#sort the dataset by price in ascending order
import pandas as pd
df1=pd.read_excel("/content/pandas_filter_sort_eda (3).xlsx")
df2=df1.sort_values(by="Price",ascending=True)
print(df2)
```

```python
#sort the dataset by quantity in descending order
import pandas as pd
df1=pd.read_excel("/content/pandas_filter_sort_eda (3).xlsx")
df2=df1.sort_values(by="Quantity",ascending=0)
print(df2)
```

```python
#sort the dataset by category(A-Z) and then by price(high-low)
import pandas as pd
df1=pd.read_excel("/content/pandas_filter_sort_eda (3).xlsx")
df2=df1.sort_values(by=["Category","Price"],ascending=[True,False])
print(df2)
```

```python
#find the top 10 highest priced order
import pandas as pd
df1=pd.read_excel("/content/pandas_filter_sort_eda (3).xlsx")
df2=df1.sort_values(by="Price",ascending =False)
print(df2.head(10))
```
