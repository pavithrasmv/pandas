```python
#find tot revenue generated in each region
import pandas as pd
df1=pd.read_excel("/content/groupby_eda_data.xlsx")

grouped=df1.groupby("Region")["Revenue"]
result=grouped.sum()
print(result)
```

```python
import pandas as pd

df1 = pd.read_excel("/content/groupby_eda_data.xlsx")

res = df1.groupby("Product")["Units_Sold"].sum()

print(res)
```

```python
#find avg revenue for each sales_person
import pandas as pd
df1=pd.read_excel("/content/groupby_eda_data.xlsx")
grouped=df1.groupby("Sales_Person")["Revenue"].mean()

print(grouped)
```

```python
#find max revenue recorded in each region
import pandas as pd
df1=pd.read_excel("/content/groupby_eda_data.xlsx")
grouped=df1.groupby("Region")["Revenue"].max()
print(grouped)
```

```python
#find minimum  units_sold for each region
import pandas as pd
df1=pd.read_excel("/content/groupby_eda_data.xlsx")
grouped=df1.groupby("Region")["Units_Sold"].min()
print(grouped)

```

```python
#count no of  orders handled by each sales_person
import pandas as pd
df1=pd.read_excel("/content/groupby_eda_data.xlsx")
grouped=df1.groupby("Sales_Person")["Order_ID"].count()
print(grouped)
```

```python
#find tot revenue for each product in each region
import pandas as pd
df1=pd.read_excel("/content/groupby_eda_data.xlsx")
cols=["Region","Product"]
grouped=df1.groupby(cols)["Revenue"]
res=grouped.sum()
print(res)
```

```python
#find avg units_sold for each product
import pandas as pd
df1=pd.read_excel("/content/groupby_eda_data.xlsx")
grouped=df1.groupby("Product")["Units_Sold"].mean()
print(grouped)
```

```python
#find top 3 products  based on tot revenue
import pandas as pd
df1=pd.read_excel("/content/groupby_eda_data.xlsx")
grouped=df1.groupby("Product")["Revenue"].sum()
print(grouped)
print()
print(grouped.tail(3))
```

```python
#find sales_person with highest tot revenue
import pandas as pd

df1 = pd.read_excel("/content/groupby_eda_data.xlsx")

res = df1.groupby("Sales_Person")["Revenue"].sum()

print(res)
print("Highest Revenue:", res.max())
print("Sales Person:", res.idxmax())
```
