```python
#giving customize name to index
import pandas as pd
values=[70,60,50,40]
marks=["tamil","maths","eng","sci"]
s=pd.Series(values,index=marks)
print(s)
```

```python
#giving customise col name and index name to Series
import pandas as pd
values=[70,60,50,40]
marks=["tamil","maths","eng","sci"]
s=pd.Series(values,index=marks,name="score")
print(s)
```

```python
#accessing index values
import pandas as pd
values=["ram","ravi","kavi","sita"]
number=[2,3,4,5]
s=pd.Series(values,index=number)
print(s)
print(s.index[2])
print(s[2])
```

```python
#accessing values
import pandas as pd
price=[1000,2000,30000]
products=["nokia","samsung","apple"]
s=pd.Series(price,index=products)
print(s)
print(s["apple"])
```
