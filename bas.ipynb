```python
#giving col names in dataframe
import pandas as pd
d=[
    ["ram",40,50],
    ["selvi",60,70],
    ["kumar",60,80],
    ["reena",70,90],
    ["karthik",60,80]
]
cols=["Name","mark1","mark2"]
df1=pd.DataFrame(d,columns=cols)
print(df1)
```

```python
#creating data frame using dictionary
import pandas as pd
d={
    "name":["ram","daniel","abhi","veeru"],
    "age":[20,20,19,18]
}
df1=pd.DataFrame(d)
print(df1)
```

```python
#creating dataframe using dictionary of dictionaries
import pandas as pd
d={
    "student1":{"name":"pavi","age":20,"grade":"A+"},
    "student2":{"name":"daniel","age":20,"grade":"B"},
    "student3":{"name":"abi","age":19,"grade":"A"}
}
df1=pd.DataFrame(d)
print(df1)
```

```python
#creating dataframe and giving index
import pandas as pd
details=[
    ["sagar",20,10000],
    ["daniel",16,20000],
    ["veeru",24,30000],
    ["raju",25,40000],
    ["kiran",26,50000]
]
col=["name","age","salary"]
i=[10,20,30,40,50]
df1=pd.DataFrame(details,columns=col,index=i)
print(df1)
```
