```python
#find product where price is between 10,000 and 30,000
import pandas as pd
df1=pd.read_excel("/content/pandas_filter_sort_eda (3).xlsx")
a=(df1["Price"]>10000)&(df1["Price"]<30000)
df2=df1[a]
print(df2)
```

```python
#filter rows where category ="Electronics " and quantity >=5
import pandas as pd
df1=pd.read_excel("/content/pandas_filter_sort_eda (3).xlsx")
a=(df1["Category"]=="Electronics")&(df1["Quantity"]>=5)
df2=df1[a]
print(df2)
```
