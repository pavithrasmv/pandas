```python
#inner join
import pandas as pd
d1={"Id":[1,2,3,4,5,6],
    "Name":["pradhan","veeru","madhu","ram","shafi","veeru"],
    "subject":["eng","Java","Html","Python","C","dotnet"]}
d2={"Id":[11,12,13,14,15,16],
    "Name":["srinu","sumanthi","neelima","daniel","arjun","veeru"],
    "subject":["Java","Html","Python","C","CPP","dotnet"]}
df1=pd.DataFrame(d1)
df2=pd.DataFrame(d2)
inn_join=pd.merge(df1,df2,on="subject",how="inner")
print(inn_join)
```

```python
import pandas as pd
d1={"Id":[1,2,3,4,5,6],
    "Name":["pradhan","veeru","madhu","ram","shafi","veeru"],
    "subject":["eng","Java","Html","Python","C","dotnet"]}
d2={"Id":[11,12,13,14,15,16],
    "Name":["srinu","sumanthi","neelima","daniel","arjun","veeru"],
    "subject":["Java","Html","Python","C","CPP","dotnet"]}
df1=pd.DataFrame(d1)
df2=pd.DataFrame(d2)
left_join=pd.merge(df1,df2,on="subject",how="left")
print(left_join)
```
