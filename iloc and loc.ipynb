```python
#select the first 5 rows using iloc
import pandas as pd
df1=pd.read_excel("/content/loc_iloc_2000_rows.xlsx")
print(df1.iloc[0:5])
```

```python
#select the last 10 rows using iloc
import pandas as pd
df1=pd.read_excel("/content/loc_iloc_2000_rows.xlsx")
print(df1.iloc[-10:])
```

```python
#select rows 100 to 110(inclusive)using iloc
import pandas as pd
df1=pd.read_excel("/content/loc_iloc_2000_rows.xlsx")
print(df1.iloc[100:111])
```

```python
#select only column "Name" using iloc
import pandas as pd
df1=pd.read_excel("/content/loc_iloc_2000_rows.xlsx")
s=df1.iloc[:,1]
print(s)
```

```python
#select age and salary using iloc
import pandas as pd
df1=pd.read_excel("/content/loc_iloc_2000_rows.xlsx")
s=df1.iloc[:,[2,4]]
print(s)
```

```python
#select y´the first 5 rows using loc
import pandas as pd
df1=pd.read_excel("/content/loc_iloc_2000_rows.xlsx")
print(df1.loc[0:4])
```

```python
#select Id,name,salary using loc
import pandas as pd
df1=pd.read_excel("/content/loc_iloc_2000_rows.xlsx")
print(df1.loc[:,["ID","Name","Salary"]])
```

```python
#select rows where department ="IT"using loc
import pandas as pd
df1=pd.read_excel("/content/loc_iloc_2000_rows.xlsx")
a=df1["Department"]=="IT"
df2=df1.loc[a]
print(df2)
```
