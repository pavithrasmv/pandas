```python
#select the first 5 rows using iloc
import pandas as pd
df1=pd.read_excel("/content/loc_iloc_2000_rows.xlsx")
print(df1.iloc[0:5])
```

```python
#select rows 100 to 110(inclusive)using iloc
import pandas as pd
df1=pd.read_excel("/content/loc_iloc_2000_rows.xlsx")
print(df1.iloc[100:111])
```

```python
#select only column "Name" using iloc
import pandas as pd
df1=pd.read_excel("/content/loc_iloc_2000_rows.xlsx")
s=df1.iloc[:,1]
print(s)
```

```python
#select age and salary using iloc
import pandas as pd
df1=pd.read_excel("/content/loc_iloc_2000_rows.xlsx")
s=df1.iloc[:,[2,4]]
print(s)
```

```python
#select y´the first 5 rows using loc
import pandas as pd
df1=pd.read_excel("/content/loc_iloc_2000_rows.xlsx")
print(df1.loc[0:4])
```

```python
#select Id,name,salary using loc
import pandas as pd
df1=pd.read_excel("/content/loc_iloc_2000_rows.xlsx")
print(df1.loc[:,["ID","Name","Salary"]])
```

```python
#select rows where department ="IT"using loc
import pandas as pd
df1=pd.read_excel("/content/loc_iloc_2000_rows.xlsx")
a=df1["Department"]=="IT"
df2=df1.loc[a]
print(df2)
```

```python
#select rows where age >50 using loc
import pandas as pd
df1=pd.read_excel("/content/loc_iloc_2000_rows.xlsx")
a=df1["Age"]>50
df2=df1.loc[a]
print(df2)
```

```python
#select rows where salary >=60000 using loc
import pandas as pd
df1=pd.read_excel("/content/loc_iloc_2000_rows.xlsx")
a=df1["Salary"]>=60000
df2=df1.loc[a]
print(df2)
```

```python
#using iloc select rows 10to 20 and columns 1 to 3
import pandas as pd
df1=pd.read_excel("/content/loc_iloc_2000_rows.xlsx")
s=df1.iloc[10:21,1:4]
print(s)
```

```python
#using loc select rows 50 to 60 and columns name and score
import pandas as pd
df1 = pd.read_excel("/content/loc_iloc_2000_rows.xlsx")
s = df1.loc[50:61, ["Name", "Score"]]
print(s)
```

```python
#select rows where Department="HR" and salary >50000
import pandas as pd
df1=pd.read_excel("/content/loc_iloc_2000_rows.xlsx")
a=(df1["Department"]=="HR") & (df1["Salary"]>50000)
df2=df1.loc[a]
print(df2)
```

```python
#select name and salary for people whose age >25
import pandas as pd
df1=pd.read_excel("/content/loc_iloc_2000_rows.xlsx")
a=(df1["Age"]>25) & (df1["Salary"])
df2=df1.loc[a]
print(df2)
```

```python
#select rows where score is between 70 and 90
import pandas as pd
df1=pd.read_excel("/content/loc_iloc_2000_rows.xlsx")
a=(df1["Score"]>70) &(df1["Score"]<90)
df2=df1.loc[a]
print(df2)
```
