```python
#concatenating one dataframe to another
import pandas as pd
d1=[[11,22],[33,44]]
d2=[[55,66],[77,88]]
c1=["A","B"]
df1=pd.DataFrame(d1,columns=c1)
df2=pd.DataFrame(d2,columns=c1)
res=[df1,df2]
df3=pd.concat(res)
print(df3)
```

```python
#ignore index
import pandas as pd
d1=[[11,22],[33,44]]
d2=[[55,66],[77,88]]
c1=["A","B"]
df1=pd.DataFrame(d1,columns=c1)
df2=pd.DataFrame(d2,columns=c1)
res=[df1,df2]
df3=pd.concat(res,ignore_index=True)
print(df3)
```

```python
#axis=0 and axis =1
import pandas as pd
d1=[[11,22],[33,44]]
d2=[[55,66],[77,88]]
c1=["A","B"]
df1=pd.DataFrame(d1,columns=c1)
df2=pd.DataFrame(d2,columns=c1)
res=[df1,df2]
df3=pd.concat(res,axis=1)
df4=pd.concat(res,axis=0)
print(df3)
print()
print(df4)
```

```python
#NaN
import pandas as pd
d1=[[11,22],[33,44]]
d2=[[55,66],[77,88],[67,99]]
c1=["A","B"]
c2=["X","Y"]
df1=pd.DataFrame(d1,columns=c1)
df2=pd.DataFrame(d2,columns=c2)
res=[df1,df2]
df3=pd.concat(res,axis=1,ignore_index=True)
df4=pd.concat(res,axis=0,ignore_index=True)
print(df3)
print()
print(df4)
```
